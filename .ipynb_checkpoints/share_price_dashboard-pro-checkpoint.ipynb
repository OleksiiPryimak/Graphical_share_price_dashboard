```python
import pandas as pd

df = pd.read_csv(
    'data\fb.csv',
    parse_dates=['Date'],
    index_col="Date",
).rename(columns={
    'Close/Last': 'close',
    'Open': 'open',
    'High': 'high',
    'Low': 'low',
    'Volume': 'volume',
}).sort_index()



# p = df.pivot_table(columns='DOW', values='Changes', aggfunc=np.mean).round(0)
# p.fillna(0, inplace=True)

# columns = ["1Sun", "2Mon", "3Tue", "4Wed", "5Thu", "6Fri", "7Sat"]
# p = p.reindex(columns=columns)
# p[columns] = p[columns].astype(int)


df
```

```python
def remove_dollar_sign(x: str) -> float:
    return float(x[1:])

df = df.assign(
    close=df['close'].map(remove_dollar_sign),
    open=df['open'].map(remove_dollar_sign),
    high=df['high'].map(remove_dollar_sign),
    low=df['low'].map(remove_dollar_sign),
)

df
```

```python
# df = df.assign(
#     close_mean = df['close'].mean()
# )
# df

df = df.assign(
    close_mean = df['close'].rolling('45D').mean(),
)
df

# df = df.assign(
#     close_mean = df['close'].rolling('60D').mean()
# )
# df
```

```python
from_ = '2021'
to = '2022'
df[from_:to].plot(
    y=['close', 'close_mean', 'volume'],
    figsize=(12, 7),
    grid=True,
    secondary_y=['volue'],
)
```
